```python

print("--- 원본 review_df ---")
print(review_df.head())
print("\n--- 원본 store_df ---")
print(store_df.head())

# 1. store_df의 '가게' 목록 (비교 대상 목록)
# 유니크한 값만 사용합니다.
store_choices = store_df['가게'].unique()

# 2. review_df의 'place_name' 목록 (매칭하려는 목록)
# 유니크한 값만 추출하여 중복 계산을 피합니다.
review_places_unique = review_df['place_name'].unique()

# 3. 각 유니크한 'place_name'에 대해 가장 유사한 '가게' 이름을 찾기
# process.extractOne은 (찾은 항목, 유사도 점수) 튜플을 반환합니다.
# 점수(score_cutoff) 임계값을 설정하여 일정 수준 이하의 매칭은 제외할 수 있습니다.
# 예: 80점 이상만 매칭
SCORE_CUTOFF = 80

# 유니크한 place_name을 기준으로 매칭 결과를 딕셔너리로 생성
# { '원본 place_name' : ('매칭된 가게 이름', 매칭 점수) }
match_mapping = {}
for place in review_places_unique:
    # process.extractOne은 가장 유사한 항목 1개를 (choice, score) 튜플로 반환
    # scorer=fuzz.WRatio는 더 정교한 비교를 위해 사용될 수 있습니다 (기본값은 fuzz.WRatio와 유사)
    match = process.extractOne(place, store_choices, scorer=fuzz.WRatio, score_cutoff=SCORE_CUTOFF)
    
    if match:
        match_mapping[place] = match  # (matched_store_name, score)
    else:
        match_mapping[place] = (None, 0) # 매칭 실패 또는 점수 미달

print(f"\n--- 유니크 place_name 기준 매칭 결과 (상위 5개) ---")
print({k: match_mapping[k] for k in list(match_mapping)[:5]})

# 4. 매칭 결과를 원래 review_df에 적용
# 딕셔너리를 map 함수에 사용하여 매칭 결과를 빠르게 적용합니다.

# 'matched_store_name' 컬럼 추가
review_df['matched_store_name'] = review_df['place_name'].map(lambda x: match_mapping.get(x, (None, 0))[0])
# 'match_score' 컬럼 추가
review_df['match_score'] = review_df['place_name'].map(lambda x: match_mapping.get(x, (None, 0))[1])

print("\n--- 매칭 결과가 추가된 review_df ---")
print(review_df)

# (선택 사항) 매칭된 결과를 store_df와 병합
# 'matched_store_name'과 store_df의 '가게' 컬럼을 기준으로 병합할 수 있습니다.
if not store_df.empty:
    merged_df = pd.merge(
        review_df,
        store_df,
        left_on='matched_store_name',
        right_on='가게',
        how='left' # review_df 기준으로 남김
    )
    print("\n--- store_df 정보까지 병합된 최종 결과 (예시) ---")
    # '가게' 컬럼이 중복되므로 정리 (예: 'id', '주소' 등 필요한 컬럼만 선택)
    if 'id' in merged_df.columns:
         print(merged_df[['place_name', 'matched_store_name', 'match_score', 'id', '주소']].head())
```
