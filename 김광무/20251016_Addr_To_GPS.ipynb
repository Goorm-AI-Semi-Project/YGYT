```python
import requests

KAKAO_REST_API_KEY = '430d403638a9ccddb70657adb3ed7196'

def convert_address_to_coordinates(address):
    """
    입력받은 주소를 WGS84 좌표계 좌표로 변환
    """

    url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + address
    header = {'Authorization': 'KakaoAK ' + KAKAO_REST_API_KEY}
 
    r = requests.get(url, headers=header)
 
    if r.status_code == 200:
        data = r.json()
        print(data)
        lng = float(r.json()["documents"][0]["address"]['x'])
        lat = float(r.json()["documents"][0]["address"]['y'])
    else:
        return None
    
    return lat, lng
```

```python
import numpy as np

store_df['X좌표'] = np.nan
store_df['Y좌표'] = np.nan
```

```python
def get_response_from_address(address):
    address_str = str(address)
    
    # 2. 문자열 처리 로직 수행
    # NaN이 str로 바뀌면 'nan'이라는 4글자 문자열이 됩니다.
    
    # 'nan' 이라는 문자열이 들어왔는지 확인하고 건너뛰는 조건 추가 (선택 사항)
    if address_str.lower() in ('nan', 'none', ''):
        return "" # 또는 None 반환    
    
    address_str = address_str.split('(')[0].strip()
    
    """
    입력받은 주소를 WGS84 좌표계 좌표로 변환
    """

    url = 'https://dapi.kakao.com/v2/local/search/address.json?query=' + address_str
    header = {'Authorization': 'KakaoAK ' + KAKAO_REST_API_KEY}
 
    r = requests.get(url, headers=header)
 
    return r
```

```python
print(get_response_from_address("서울특별시 강남구 도산대로56길 10 (청담동) 2, 3층").json())
```

```python
import pandas as pd
from tqdm import tqdm
import time
import numpy as np

for i in tqdm(range(len(store_df)), desc="DataFrame Row 처리"):
    if not np.isnan(store_df.iloc[i]['X좌표']):
        continue
    
    current_value = store_df.iloc[i]['주소']
    #print("current_value=",current_value)    
    
    res = get_response_from_address(current_value)
    
    if res == "" or not pd.notnull(res):
        print(f"주소이상: addr={current_value}")
        continue
    
    if res.status_code == 200:
        data = res.json()
        #print(data)
        try:
            lng = float(res.json()["documents"][0]["address"]['x'])
            lat = float(res.json()["documents"][0]["address"]['y'])
            
            store_df.loc[i, 'X좌표'] = lng
            store_df.loc[i, 'Y좌표'] = lat
        except:
            # 네트워크 등 요청 자체의 오류 처리
            print(f"요청 중 오류 발생: addr={current_value},data={data}")
    else:
        data = res.json()
        print(f"addr={current_value} is fault - data={data}")

    time.sleep(0.2)

    
```

```python
nan_count = store_df['X좌표'].isnull().sum()

# 2) NaN이 아닌 (유효) 것의 개수 확인
not_nan_count_1 = store_df['X좌표'].notnull().sum()
# 또는 간단하게 .count() 메서드 사용 (NaN을 세지 않음)
not_nan_count_2 = store_df['X좌표'].count()

# --- 3. 결과 출력 ---

print(f"--- 컬럼 분석 결과 ---")
print(f"1. 결측치(NaN) 개수: {nan_count} 개")
print(f"2. 결측치가 아닌(유효) 개수: {not_nan_count_1} 개 (또는 {not_nan_count_2} 개)")
```
