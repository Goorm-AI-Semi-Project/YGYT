```python
import pandas as pd
```

```python
store_df = pd.read_csv('./20251016_서울시_음식점_목록_GPS.csv')
```

```python
store_df
```

```python
y_diff = 0.0044915
x_diff = 0.0056619

```

```python
x_test = 127.0435776
y_test = 37.52437563
```

```python
def get_nearby_store_list(store_df, x_pos, y_pos):
    min_y = y_pos - y_diff
    max_y = y_pos + y_diff
    min_x = x_pos - x_diff
    max_x = x_pos + x_diff

    condition = ( 
        (store_df['Y좌표'] >= min_y) & 
        (store_df['Y좌표'] <= max_y) & 
        (store_df['X좌표'] >= min_x) & 
        (store_df['X좌표'] <= max_x) 
    )
                 
    result_df = store_df[condition]

    return result_df
```

```python
%%time
result_store_df = get_nearby_store_list(store_df, x_test, y_test)
result_store_df
```

```python
food_df = pd.read_csv('./20251017_TOTAL_MENU.csv')
food_df
```

```python
def food_list_from_store(store_list_df, food_list_df):
    merged_df = pd.merge(store_list_df, food_list_df, left_on='id', right_on='식당ID')

    return merged_df
```

```python
%%time
result_food_df = food_list_from_store(result_store_df, food_df)
result_food_df 
```
